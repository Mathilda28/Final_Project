## Penjelasan Dataset

**Fitur**

- age (numeric)
- job : type of job (categorical: "admin.","unknown","unemployed","management","housemaid","entrepreneur","student","blue-collar","self-employed","retired","technician","services")
- marital : marital status (categorical: "married","divorced","single"; note: "divorced" meansdivorced or widowed)
- education (categorical: "unknown","secondary","primary","tertiary")
- default: has credit in default? (binary: "yes","no")
- balance: average yearly balance, in euros (numeric)
- housing: has housing loan? (binary: "yes","no")
- loan: has personal loan? (binary: "yes","no")
- contact: contact communication type (categorical: "unknown","telephone","cellular")
- day: last contact day of the month (numeric)
- month: last contact month of year (categorical: "jan", "feb", "mar", ..., "nov", "dec")
- duration: last contact duration, in seconds (numeric)
- campaign: number of contacts performed during this campaign and for this client (numeric,includes last contact)
- pdays: number of days that passed by after the client was last contacted from a previouscampaign (numeric, -1 means client was not previously contacted)
- previous: number of contacts performed before this campaign and for this client (numeric)
- poutcome: outcome of the previous marketing campaign
(categorical:"unknown","other","failure","success")

**Target**
- Deposit

  yes = nasabah menjadi nasabah deposit

  no = nasabah menolak deposit

### VIF

```python
#VIF dengan data numerik
from statsmodels.stats.outliers_influence import variance_inflation_factor
def vifcalculation(data):

    # menghitung VIF
    vif = pd.DataFrame()
    vif["variables"] = data.columns
    vif["VIF"] = [variance_inflation_factor(data.values, i) for i in range(data.shape[1])]

    return(vif)
```


VIF (Variance Inflation Factor) adalah alat statistik yang digunakan dalam analisis regresi untuk mengukur tingkat multikolinearitas antara fitur atau variabel independen dalam model. Multikolinearitas terjadi ketika dua atau lebih variabel independen dalam model regresi memiliki hubungan yang kuat satu sama lain, yang dapat mempengaruhi interpretasi dan stabilitas model.

```python
#VIF Calculation
vifcalculation(num)
```

### Feature Scaling

```python
# Encoding on categorical pipe
cat_pipe = Pipeline([
        ('cat_ohe', OneHotEncoder(sparse=False, handle_unknown='ignore'))
    ])

# Create numerical pipeline
num_pipeline = Pipeline([
        ('outlier', Winsorizer(capping_method='iqr', tail='both', fold=1.5)),
        ('imputer', SimpleImputer(strategy='median')),
        ('scaler', MinMaxScaler())
    ])

# Combine pipelines using ColumnTransformer
full_pipe = ColumnTransformer([
        ("numcol", num_pipeline, ['duration', 'pdays', 'day', 'age', 'campaign']),
        ("catcol", cat_pipe, ['housing', 'month', 'loan', 'poutcome', 'job'])])


```

Hal ini menghasilkan akurasi yang tinggi dan kemampuan untuk menangani interaksi yang kompleks antara fitur-fitur.
Dalam memilih model terbaik, evaluasi kinerja melalui cross-validation dan pemahaman mendalam tentang karakteristik data adalah kunci. Masing-masing model memiliki trade-off antara kompleksitas dan akurasi, serta fitur-fitur khusus yang dapat mendukung tugas klasifikasi customer deposit.

5. K-Nearest Neighbors (KNN):

Alasan : KNN adalah model yang sederhana dan intuitif untuk klasifikasi. Model ini dapat digunakan untuk memprediksi apakah seorang nasabah akan berlangganan deposito berjangka atau tidak. Namun, penting untuk mempertimbangkan kelemahan dan kelebihan KNN serta melakukan evaluasi kinerja untuk memastikan model memberikan prediksi yang akurat dan sesuai dengan tujuan bisnis.

```python
pipeline_rf = Pipeline([
    ('preprocess', full_pipe),
    ('model', RandomForestClassifier(random_state=42))
])

pipeline_dectree = Pipeline([
    ('preprocess', full_pipe),
    ('model', DecisionTreeClassifier(random_state=42))
])

pipeline_nb = Pipeline([
    ('preprocess', full_pipe),
    ('model', GaussianNB())
])

pipeline_gb = Pipeline([
    ('preprocess', full_pipe),
    ('model', GradientBoostingClassifier(random_state=42))
])

pipeline_knn = Pipeline([
    ('preprocess', full_pipe),
    ('model', KNeighborsClassifier())
])
```
